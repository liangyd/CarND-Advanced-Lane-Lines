## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.

## First, I'll compute the camera calibration using chessboard images

```python
import numpy as np
import cv2
import glob
import matplotlib.pyplot as plt

# prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3), np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)


# Arrays to store object points and image points from all the images.
objpoints = [] # 3d points in real world space
imgpoints = [] # 2d points in image plane.

# Make a list of calibration images
images = glob.glob('camera_cal/calibration*.jpg')
# Step through the list and search for chessboard corners
for fname in images:
    img = cv2.imread(fname)
    pname= fname.split('\\')[-1]
    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6), None)
    #print(ret)
    
    # If found, add object points, image points
    if ret == True:
        objpoints.append(objp)
        imgpoints.append(corners)

        # Draw and display the corners
        img = cv2.drawChessboardCorners(img, (9,6), corners, ret)
        #cv2.imshow('img',img)
        #cv2.waitKey(500)
        # Save the chessboard images
        cv2.imwrite('output_images/chessboard/'+pname, img)

# cv2.destroyAllWindows()
# print(np.shape(images))
# print(np.shape(objpoints))

# Compute camera calibration matrix and distortion coefficients
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1], None, None)
```

## Then, I'll create a pipeline to create a binary image using color tranform and gradients

```python
# Define the function to convert a raw image to a binary image
def binary_image(img, b_thresh=(190, 255), l_thresh=(210,255)):
    
    img = np.copy(img)
    # Convert to HSV color space and separate the V channel
#     hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
#     v_channel = hsv[:,:,2]
#     v_channel = np.uint8(v_channel*(255/np.max(v_channel)))
    # Seperate the r channel in RGB color space
#     rgb = img
#     r_channel = rgb[:,:,0]
#     r_channel = np.uint8(r_channel*(255/np.max(r_channel)))
    # Convert to LAB color space and seperate the B channel
    lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
    b_channel = lab[:,:,2]
    if np.max(b_channel) > 175:
        b_channel = np.uint8(b_channel*(255/np.max(b_channel)))
    # Convert to HLS color space and seperate the L channel
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    l_channel = hls[:,:,1]
    l_channel = np.uint8(l_channel*(255/np.max(l_channel)))


    # Threshold color channel
#     v_binary = np.zeros_like(v_channel)
#     v_binary[(v_channel >= v_thresh[0]) & (v_channel <= v_thresh[1])] = 1
#     r_binary = np.zeros_like(r_channel)
#     r_binary[(r_channel >= r_thresh[0]) & (r_channel <= r_thresh[1])] = 1
    yellow_line = np.zeros_like(b_channel)
    yellow_line[(b_channel >= b_thresh[0]) & (b_channel <= b_thresh[1])] = 1
    white_line = np.zeros_like(l_channel)
    white_line[(l_channel >= l_thresh[0]) & (l_channel <= l_thresh[1])] = 1
    


    # Combined binary image using both color detection and sobel operator
    combined_binary = np.zeros_like(yellow_line)
    combined_binary[(yellow_line == 1)|(white_line==1)] = 255
    return combined_binary
```

```python
# Define the perspective transform function
def perspective_transform(img):
    img_size=(img.shape[1], img.shape[0])
    # define the src and dst array
    src = np.float32(
                [[570, 470],
                [210, 720],
                [1110, 720],
                [722, 470]])
    dst = np.float32(
                [[(img_size[0] / 4), 0],
                [(img_size[0] / 4), img_size[1]],
                [(img_size[0] * 3 / 4), img_size[1]],
                [(img_size[0] * 3 / 4), 0]])
    # calculate the transform matrix
    M = cv2.getPerspectiveTransform(src, dst)
    warped = cv2.warpPerspective(img, M, img_size)
    return warped
```

```python
# Define inverse perspective transform
def inv_perspective_transform(img):
    img_size=(img.shape[1], img.shape[0])
    # define the src and dst array
    src = np.float32(
                [[570, 470],
                [210, 720],
                [1110, 720],
                [722, 470]])
    dst = np.float32(
                [[(img_size[0] / 4), 0],
                [(img_size[0] / 4), img_size[1]],
                [(img_size[0] * 3 / 4), img_size[1]],
                [(img_size[0] * 3 / 4), 0]])
    # calculate the transform matrix
    Minv = cv2.getPerspectiveTransform(dst, src)
    unwarped = cv2.warpPerspective(img, Minv, img_size)
    return unwarped
```

## Then, I will find the lane boundary on a test image

```python
# Read a test image and convert it to binary image
img=cv2.imread('output_images/undistorted_image/test3.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
binary=binary_image(img)
# Apply perspective transform
binary_warped=perspective_transform(binary)

# Take a histogram of the bottom half of the image
histogram=np.sum(binary_warped[np.int(binary_warped.shape[0]/2):,:], axis=0)

# Create an output image to visualize the result
out_img=np.dstack((binary_warped,binary_warped,binary_warped))*255

# Find the peak of the left and right halves of the histogram
midpoint=np.int(histogram.shape[0]/2)
leftx_base=np.argmax(histogram[:midpoint])
rightx_base=np.argmax(histogram[midpoint:])+midpoint

# Set the number of the sliding window
nwindow=9
# Set the height of windows
window_height=np.int(binary_warped.shape[0]/nwindow)

#Identify the x and y positions of all nonzero pixels
nonzero=binary_warped.nonzero()
nonzeroy=np.array(nonzero[0])
nonzerox=np.array(nonzero[1])

# Current positions to be updated for each window
leftx_current=leftx_base
rightx_current=rightx_base

# Set the width of the window
margin=70
# Set the min number of pixels to recenter the window
minpix=40

#Create empty list to receive left and right lane pixel indices
left_lane_inds=[]
right_lane_inds=[]


# Step through the windows one by one
for window in range(nwindow):
    #Identify window boundaries 
    win_y_low = binary_warped.shape[0]-(window+1)*window_height
    win_y_high = binary_warped.shape[0]-window*window_height
    win_xleft_low = leftx_current-margin
    win_xleft_high = leftx_current+margin
    win_xright_low = rightx_current-margin
    win_xright_high = rightx_current+margin

    #Draw the windows on the visualization image
    cv2.rectangle(out_img, (win_xleft_low, win_y_low), (win_xleft_high, win_y_high), (255,255,0), 5)
    cv2.rectangle(out_img, (win_xright_low, win_y_low), (win_xright_high, win_y_high), (255,255,0), 5)

    #Identify the pixels within the window
    good_left_inds=((nonzeroy>=win_y_low)&(nonzeroy<=win_y_high)&(nonzerox>=win_xleft_low)&(nonzerox<win_xleft_high)).nonzero()[0]
    good_right_inds=((nonzeroy>=win_y_low)&(nonzeroy<=win_y_high)&(nonzerox>=win_xright_low)&(nonzerox<win_xright_high)).nonzero()[0]

    #Append the indices
    left_lane_inds.append(good_left_inds)
    right_lane_inds.append(good_right_inds)

    #Recenter the next window on the mean position when many pixels are found 
    if len(good_left_inds)>minpix:
        leftx_current=np.int(np.mean(nonzerox[good_left_inds]))
    if len(good_right_inds)>minpix:
        rightx_current=np.int(np.mean(nonzerox[good_right_inds]))

# Concatenate the arrays of indices
left_lane_inds=np.concatenate(left_lane_inds)
right_lane_inds=np.concatenate(right_lane_inds)

# Extract left and right line pixel positions
leftx=nonzerox[left_lane_inds]
lefty=nonzeroy[left_lane_inds]
rightx=nonzerox[right_lane_inds]
righty=nonzeroy[right_lane_inds]

# Fit a second order polynomial to each
left_fit = np.polyfit(lefty, leftx, 2)
right_fit = np.polyfit(righty, rightx, 2)


```

```python
# Define the lane finding process in a function
def lanefinder(img, left_line, right_line): 
    binary=binary_image(img)
    # Apply perspective transform
    binary_warped=perspective_transform(binary)
    
    #Identify the x and y positions of all nonzero pixels
    nonzero=binary_warped.nonzero()
    nonzeroy=np.array(nonzero[0])
    nonzerox=np.array(nonzero[1])
    
    # Set the width of the window
    margin=70
    # Set the min number of pixels to recenter the window
    minpix=40
    
    if left_line.detected == False or right_line.detected == False:
        # Take a histogram of the bottom half of the image
        histogram=np.sum(binary_warped[np.int(binary_warped.shape[0]*2/3):,:], axis=0)

        # Create an output image to visualize the result
        #out_img=np.dstack((binary_warped,binary_warped,binary_warped))*255

        # Find the peak of the left and right halves of the histogram
        midpoint=np.int(histogram.shape[0]/2)
        leftx_base=np.argmax(histogram[:midpoint])
        rightx_base=np.argmax(histogram[midpoint:])+midpoint

        # Set the number of the sliding window
        nwindow=9
        # Set the height of windows
        window_height=np.int(binary_warped.shape[0]/nwindow)

        # Current positions to be updated for each window
        leftx_current=leftx_base
        rightx_current=rightx_base

        #Create empty list to receive left and right lane pixel indices
        left_lane_inds=[]
        right_lane_inds=[]


        # Step through the windows one by one
        for window in range(nwindow):
            #Identify window boundaries 
            win_y_low = binary_warped.shape[0]-(window+1)*window_height
            win_y_high = binary_warped.shape[0]-window*window_height
            win_xleft_low = leftx_current-margin
            win_xleft_high = leftx_current+margin
            win_xright_low = rightx_current-margin
            win_xright_high = rightx_current+margin

            #Draw the windows on the visualization image
            #cv2.rectangle(out_img, (win_xleft_low, win_y_low), (win_xleft_high, win_y_high), (255,255,0), 5)
            #cv2.rectangle(out_img, (win_xright_low, win_y_low), (win_xright_high, win_y_high), (255,255,0), 5)

            #Identify the pixels within the window
            good_left_inds=((nonzeroy>=win_y_low)&(nonzeroy<=win_y_high)&(nonzerox>=win_xleft_low)&(nonzerox<win_xleft_high)).nonzero()[0]
            good_right_inds=((nonzeroy>=win_y_low)&(nonzeroy<=win_y_high)&(nonzerox>=win_xright_low)&(nonzerox<win_xright_high)).nonzero()[0]

            #Append the indices
            left_lane_inds.append(good_left_inds)
            right_lane_inds.append(good_right_inds)

            #Recenter the next window on the mean position when many pixels are found 
            if len(good_left_inds)>minpix:
                leftx_current=np.int(np.mean(nonzerox[good_left_inds]))
            if len(good_right_inds)>minpix:
                rightx_current=np.int(np.mean(nonzerox[good_right_inds]))
        
        # Concatenate the arrays of indices
        left_lane_inds=np.concatenate(left_lane_inds)
        right_lane_inds=np.concatenate(right_lane_inds)
        
        #Set the line detection status
        left_line.detected=True
        right_line.detected=True
        
    else:
        left_fit=left_line.current_fit
        right_fit=right_line.current_fit
        left_lane_inds = ((nonzerox > (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] - margin)) & (nonzerox < (left_fit[0]*(nonzeroy**2) + left_fit[1]*nonzeroy + left_fit[2] + margin))) 
        right_lane_inds = ((nonzerox > (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] - margin)) & (nonzerox < (right_fit[0]*(nonzeroy**2) + right_fit[1]*nonzeroy + right_fit[2] + margin)))  
        
    # Extract left and right line pixel positions
    left_line.allx=nonzerox[left_lane_inds]
    left_line.ally=nonzeroy[left_lane_inds]
    right_line.allx=nonzerox[right_lane_inds]
    right_line.ally=nonzeroy[right_lane_inds]
    
    # Convert left and right line pixel positions in meters
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    ym_per_pix = 30/720  # meters per pixel in y dimension
    left_line.allx_cr=nonzerox[left_lane_inds]*xm_per_pix
    left_line.ally_cr=nonzeroy[left_lane_inds]*ym_per_pix
    right_line.allx_cr=nonzerox[right_lane_inds]*xm_per_pix
    right_line.ally_cr=nonzeroy[right_lane_inds]*ym_per_pix
    
    # Fit a second order polynomial to each
    if len(nonzerox[left_lane_inds])>0:
        left_line.current_fit = np.polyfit(left_line.ally, left_line.allx, 2)
        left_line.current_fit_cr = np.polyfit(left_line.ally_cr, left_line.allx_cr, 2) 
    else:
        left_line.current_fit = left_line.recent_5fit[-1]
        left_line.current_fit_cr = left_line.recent_5fit_cr[-1]
        
    if len(nonzerox[right_lane_inds])>0:    
        right_line.current_fit = np.polyfit(right_line.ally, right_line.allx, 2)
        right_line.current_fit_cr = np.polyfit(right_line.ally_cr, right_line.allx_cr, 2)
    else:
        right_line.current_fit = right_line.recent_5fit[-1]
        right_line.current_fit_cr = right_line.recent_5fit_cr[-1]
    
    # Sanity Check: check if the two lanes are almost parrallel to each other
    # Calculate the lane distance at the top of the image
    y_eval=0
    left_fitx_top = left_line.current_fit[2]
    right_fitx_top = right_line.current_fit[2]
    diff_top=(right_fitx_top-left_fitx_top)*xm_per_pix
    # Calculate the lane distance at the bottom of the image
    y_eval=719*ym_per_pix
    left_fitx_bottom = left_line.current_fit_cr[0]*y_eval**2 + left_line.current_fit_cr[1]*y_eval + left_line.current_fit_cr[2]
    right_fitx_bottom = right_line.current_fit_cr[0]*y_eval**2 + right_line.current_fit_cr[1]*y_eval + right_line.current_fit_cr[2]
    diff_bottom=right_fitx_bottom-left_fitx_bottom
    # Calculate the difference between the current polyfit coefficients and the recent ones
    if len(right_line.recent_5fit)>0:
        right_line.diffs=abs(right_line.current_fit-right_line.recent_5fit[-1])
    if len(left_line.recent_5fit)>0:
        left_line.diffs=abs(left_line.current_fit-left_line.recent_5fit[-1])
    
    histogram=np.sum(binary_warped[np.int(binary_warped.shape[0]*9/10):,:], axis=0)
    
    # Reset the sliding window and reject the polyfit coefficients if the sanity check fails
    if diff_top<1.5 or diff_top>4 or diff_bottom<1.5 or diff_bottom>4 or abs(diff_bottom-diff_top)>1.5 or max(histogram)<500:
        left_line.current_fit = left_line.recent_5fit[-1]
        left_line.current_fit_cr = left_line.recent_5fit_cr[-1]
        right_line.current_fit = right_line.recent_5fit[-1]
        right_line.current_fit_cr = right_line.recent_5fit_cr[-1]
        #Reset the line detection status
        left_line.detected=False
        right_line.detected=False
        
    if right_line.diffs[0]>0.001 or right_line.diffs[1]>1.0 or right_line.diffs[2]>100.0:
        right_line.current_fit = right_line.recent_5fit[-1]
        right_line.current_fit_cr = right_line.recent_5fit_cr[-1]
        #Reset the line detection status
        left_line.detected=False
        right_line.detected=False
        
    if left_line.diffs[0]>0.001 or left_line.diffs[1]>1.0 or left_line.diffs[2]>100.0:
        left_line.current_fit = left_line.recent_5fit[-1]
        left_line.current_fit_cr = left_line.recent_5fit_cr[-1]
        #Reset the line detection status
        left_line.detected=False
        right_line.detected=False
        
```

## After that, I will determine the curvature of the lane and vehicle position with respect to center.

```python
def curvature(left_fit_cr, right_fit_cr):
    
    y_eval = 719
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    ym_per_pix = 30/720  # meters per pixel in y dimension
    
#     leftx=left_line.allx
#     lefty=left_line.ally
#     rightx=right_line.allx
#     righty=right_line.ally
#     # Fit new polynomials to x,y in world space
#     left_fit_cr = np.polyfit(lefty*ym_per_pix, leftx*xm_per_pix, 2)
#     right_fit_cr = np.polyfit(righty*ym_per_pix, rightx*xm_per_pix, 2)
#     left_fit_cr=left_line.current_fit_cr
#     right_fit_cr=right_line.current_fit_cr
    # Calculate the new radii of curvature
    left_curverad = ((1 + (2*left_fit_cr[0]*y_eval*ym_per_pix + left_fit_cr[1])**2)**1.5) / np.absolute(2*left_fit_cr[0])
    right_curverad = ((1 + (2*right_fit_cr[0]*y_eval*ym_per_pix + right_fit_cr[1])**2)**1.5) / np.absolute(2*right_fit_cr[0])
    curverad=(left_curverad+right_curverad)/2
    return curverad
    
```

```python
def offset(left_fit, right_fit):
    xm_per_pix = 3.7/700 # meters per pixel in x dimension
    # Calculate the intersection between the lane lines and the bottom of the image
    y_eval=719
    left_fitx = left_fit[0]*y_eval**2 + left_fit[1]*y_eval + left_fit[2]
    right_fitx = right_fit[0]*y_eval**2 + right_fit[1]*y_eval + right_fit[2]
    # Calculate vehicle center and the lane center
    lane_center=(left_fitx+right_fitx)/2
    vehicle_center=img.shape[1]/2
    # Calculate the offset (if positive, the vehicle is close to the right lane, otherwise, it is to the left)
    center_offset_pixels = vehicle_center - lane_center
    return center_offset_pixels*xm_per_pix
```

## Draw the detected lane boundaries back onto the original image.

```python
def draw(img, left_fit, right_fit, curvature, vehicle_position_offset):
    # Get all the pixels on the left and right lanes
    ploty = np.linspace(0, img.shape[0]-1, img.shape[0] )
    left_fitx = left_fit[0]*ploty**2 + left_fit[1]*ploty + left_fit[2]
    right_fitx = right_fit[0]*ploty**2 + right_fit[1]*ploty + right_fit[2]

    # Create an image to draw the lines on
    color_warp = np.zeros_like(img)

    # Recast the x and y points into usable format for cv2.fillPoly()
    pts_left = np.array([np.transpose(np.vstack([left_fitx, ploty]))])
    pts_right = np.array([np.flipud(np.transpose(np.vstack([right_fitx, ploty])))])
    pts = np.hstack((pts_left, pts_right))

    # Draw the lane onto the warped blank image
    cv2.fillPoly(color_warp, np.int_([pts]), (0,255, 0))

    # Warp the blank back to original image space using inverse perspective matrix
    newwarp = inv_perspective_transform(color_warp)
    # Combine the result with the original image
    result = cv2.addWeighted(img, 1, newwarp, 0.3, 0)

    # Put curvature information onto the image
    font = cv2.FONT_HERSHEY_DUPLEX
    text = 'Radius of Curvature = ' + '{:04.2f}'.format(curvature) + ' m'
    cv2.putText(result, text, (40,70), font, 1.5, (200,255,155), 2, cv2.LINE_AA)

    # Put the offset information onto the image
    direction = ''
    if vehicle_position_offset > 0:
        direction = 'right'
    elif vehicle_position_offset < 0:
        direction = 'left'
    abs_offset = abs(vehicle_position_offset)
    text = 'Vehicle is '+'{:04.3f}'.format(abs_offset) + ' m ' + direction + ' of center'
    cv2.putText(result, text, (40,120), font, 1.5, (200,255,155), 2, cv2.LINE_AA)
    return result
```

## Define the complete image processing pipeline

```python
# Define a class to receive the characteristics of each line detection
class Line():
    def __init__(self):
        # was the line detected in the last iteration?
        self.detected = False   
        #polynomial coefficients for the most recent fit
        self.current_fit = [] 
        #polynomial coefficients for the most recent fit in real size
        self.current_fit_cr = [] 
        #polynomial coefficients for the last 5 fits of the line
        self.recent_5fit = [] 
        #polynomial coefficients for the last 5 fits of the line in real size
        self.recent_5fit_cr = []   
        #polynomial coefficients averaged with the last 5 fits
        self.best_fit = None  
        #polynomial coefficients averaged with the last 5 fits in real size
        self.best_fit_cr = None  
        #x values for detected line pixels
        self.allx = None  
        #y values for detected line pixels
        self.ally = None
        #difference in fit coefficients between last and new fits
        self.diffs = np.array([0,0,0], dtype='float') 
```

```python
def smooth(line):
    
    if len(line.recent_5fit)<5:
        line.recent_5fit.append(line.current_fit)
        line.recent_5fit_cr.append(line.current_fit_cr)
        # the best fit in pixel
        line.best_fit=line.current_fit
        # the best fit in real size
        line.best_fit_cr=line.current_fit_cr
    else:
        # the best fit in pixel
        line.recent_5fit.pop(0)
        line.recent_5fit.append(line.current_fit)
        line.best_fit=np.sum(line.recent_5fit, axis=0)/5
        # the best fit in real size
        line.recent_5fit_cr.pop(0)
        line.recent_5fit_cr.append(line.current_fit_cr)
        line.best_fit_cr=np.sum(line.recent_5fit_cr, axis=0)/5
```

```python
def img_process(img):
    # Distortion Correction
    undist = cv2.undistort(img, mtx, dist, None, mtx)
    # Get the pixels on both lane lines
    lanefinder(undist, left_line, right_line)
    
    #Smooth
    smooth(left_line)
    smooth(right_line)
    # Calculate curvature and offset
    curverad=curvature(left_line.best_fit_cr, right_line.best_fit_cr)
    vehicle_position_offset=offset(left_line.best_fit, right_line.best_fit)
    
    # Draw the lines and information on the image
    result=draw(img, left_line.best_fit, right_line.best_fit, curverad, vehicle_position_offset)
    return result
```
